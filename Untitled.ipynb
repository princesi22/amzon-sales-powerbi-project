```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('blinkit_data.csv')
```

```python
df.head(10)
```

# data cleaning


```python
df['Item Fat Content']= df['Item Fat Content'].replace({'LF':'Low Fat','reg':'Regular','low fat':'Low Fat'})
```

```python
print(df['Item Fat Content'].unique())
```

# KPIs


# Total Sales

```python
total_sales = df['Sales'].sum()
```

# Avg Sales

```python
avg_sales = df['Sales'].mean()
```

# Number of items Sold


```python
number_of_item_sold = df['Item Type'].count()
```

# Avg Ratings

```python
avg_ratings = df['Rating'].mean()
```

# Display

```python
print(f"Total Sales : ${total_sales:,.1f}")
```

```python
print(f"Avg Ratings : {avg_ratings:.0f}")
```

```python
print(f"Number Of Items Sold : {number_of_item_sold:,.0f}")
```

```python
print (f"Avg Sales : ${avg_sales:,.1f}")
```

# CHARTS

# total sales by fat content

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie (sales_by_fat , labels = sales_by_fat.index,
         autopct = '%.1f%%',
         startangle = 90)
plt.title ('sales by fat content')
plt.axis('equal')
plt.show()
```

# total sales by item

```python
import matplotlib.pyplot as plt

sales_by_item = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_item.index, sales_by_item.values)  # Use index for labels

plt.xticks(rotation=-90)
plt.xlabel('Item Type')  # corrected
plt.ylabel('Total Sales')  # corrected
plt.title("Total Sales by Item")
plt.tight_layout()  # optional, helps fit rotated labels
plt.show()
```


# fat content by outlet for total sales 

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]  # reorder columns if needed

ax = grouped.plot(kind='bar', figsize=(8, 5), title="Outlet Tier by Item Fat Content")

plt.xlabel('Outlet Location Type')  # no indentation
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

# total sale by outlet establishment

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize= (9,5))
plt.plot (sales_by_year.index, sales_by_year.values,
          marker = 'o', linestyle = '-')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title("total sales bu outlet establishment")
plt.tight_layout()
plt.show()
```

# sales by outlet size

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(4,4))
plt.pie (sales_by_size , labels= sales_by_size.index,
         autopct = '%.1f%%', startangle = 90)
plt.title("sales by outlet by size")
plt.tight_layout()
plt.show()
```

# Sales by outlet Location

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales',ascending=False)
plt.figure(figsize=(8,3))
sns.barplot (x='Sales', y='Outlet Location Type', data = sales_by_location)
plt.title ('Total sales by Outlet Location Type')
plt.xlabel ('Total Sales')
plt.tight_layout()
plt.show()
```
